## 5.- Crear un nuevo dataframe solo con columnas de valor numerico , crea una matriz de correlación, elimina la columna del valor nulo (2pts)

```python

""" Creacion de una variable que contiene las columnas con valores numericos del df original
    Creacion de un df que contenga las columnas numericas del df original
    Creacion de la matriz de correlacion
    Eliminamos las columnas con datos nulos
    Eliminamos las filas con datos nulos"""
    
columnas_numericas = df.columns[df.dtypes != 'object']
df_numerico = df.loc[:, columnas_numericas]
matriz_correlacion = df_numerico.corr()
matriz_correlacion.dropna(axis=1, thresh=2, inplace=True)
matriz_correlacion.dropna(axis=0, thresh=2, inplace=True)
print(matriz_correlacion)
```

## 6.- En la Matriz anterior que columna tiene una correlación negativa. Valor (2ps)

```python
"""un programa que retorna la columna que tiene mas de 1 valor negativo,
ya que en la matriz, todas las columnas tendran la fila rating con un valor negativo"""
for column in matriz_correlacion:
    contador = 0
    for value in matriz_correlacion[column]:
        if value < 0:
            contador += 1
    if contador >1:
        print(column)
        break
```

## 8.- Crear un DataFrame solo con las variables categoricas y pasar todos los valores a mayuscula. Valor (2pts)

```python
"""Extraccion de las columnas del df original que sean object, y almacenadas en columnas_categoricas
    Creamos el df_categorico con esas columnas
    aplicamos una funcion lambda a cada columna, que pase a mayusculas el valor"""

columnas_categoricas = df.columns[df.dtypes == 'object']
df_categorico = df.loc[:, columnas_categoricas]
df_categorico = df_categorico.apply(lambda x: x.str.upper())

df_categorico.head()
```

## 10.- Traducir todas los nombres de columnas al español.  Valor (2pts)

```python
"""Instalacion e importacion del modulo googetrans"""
from googletrans import Translator 
translator = Translator()
```

```python
"""Creamos una lista que contenga las columnas del df"""

list_Trans=[] 
for elemento in df.columns:
    list_Trans.append (elemento)
list_Trans[:3]
```

```python
"""Creamos una lista2 que contenga las columnas de lista1
    Traducidas al espanol"""

list_Trans2=[]
for elemento in list_Trans:
    translation=translator.translate(elemento, src='en', dest='es')
    list_Trans2.append(translation.text)
list_Trans2[:3]
```

```python
"""Modificamos cada columna del df_concatenado
con un for que las recorra y les cambie el valor
a los de la lista traducida con el mismo indice"""

for i in range(len(list_Trans)):
    df_concatenado.rename(columns={f'{list_Trans[i]}':f'{list_Trans2[i]}'}, inplace=True)
```

## 11.- Crear una función que retorne si los valores de la columna cantidad son par o impar, guardar el resultado en una nueva columna. Valor (5pts)

```python
"""Creamos una funcion que retorne si un numero es par o impar"""

def par(n):
    if n%2 == 0:
        return 'Par'
    else:
        return 'Impar'
        
df_concatenado['Par'] = df_concatenado.Cantidad.apply(par) 
"""Creamos una columna llamada par, que contenga los valores de cantidad con la funcion par() aplicada"""
```

## 12.- Sumar todos los valores de la columna total donde el valor de cantidad es par, y restarlo al impar. Valor (5pts)

```python
"""Creacion de dos variables que almacenan la suma de los
valores de la columna cantidad que sean pares, e impares
hace la cuenta y la retorna"""
suma_par = df_concatenado[df_concatenado.Cantidad % 2 == 0].Total.sum()
suma_impar = df_concatenado[df_concatenado.Cantidad % 2 != 0].Total.sum()
print(suma_impar - suma_par)
```
